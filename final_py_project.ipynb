# **Customer Analysis**

Most valuable Customers

```python
valuable_customers = df.groupby('CustomerKey')['Total_sales'].sum().sort_values(ascending=False).reset_index().head(10)
valuable_customers['Cumulative_pctg'] = ((valuable_customers['Total_sales'].cumsum())/(valuable_customers['Total_sales'].sum()))*100
valuable_customers.shape
```

```python
from matplotlib.ticker import PercentFormatter
fig, ax1 = plt.subplots(figsize=(12,8))
sns.barplot(x = 'CustomerKey', y = 'Total_sales', data=valuable_customers, color='skyblue')
ax1 = plt.gca().twinx()
sns.lineplot(x = 'CustomerKey', y='Cumulative_pctg', data=valuable_customers, color = 'red')
ax1.yaxis.set_major_formatter(PercentFormatter())
plt.title('Pareto Chart of Customer Spending')
plt.xlabel('CustomerKey')
plt.ylabel('Total Spending')
plt.xticks(rotation=30)
plt.show()
```

What are the Price Elasticities of Demand for Products

```python
product_sales = df.groupby(['ProductKey', 'Months'])['Total_sales'].sum().reset_index()
product_sales['SalesChange'] = product_sales.groupby('ProductKey')['Total_sales'].pct_change()*100
```

```python
elasticity = df.groupby(['ProductKey', 'Months']).agg(
                                                   total_price =('Price','sum'),
                                                    total_quantity  =('Qty', 'sum')).reset_index()
elasticity['price_change'] = elasticity.groupby('ProductKey')['total_price'].pct_change()*100
elasticity['quantity_change'] = elasticity.groupby('ProductKey')['total_quantity'].pct_change()*100
elasticity['Elasticity'] = elasticity['quantity_change']/elasticity['price_change']
elasticity
```

```python
plt.bar(elasticity['ProductKey'], elasticity['Elasticity'])
plt.title('Price Elasticity of Demand Over Time')
plt.xlabel('Product Key')
plt.ylabel('Elasticity')
plt.xticks(rotation = 45)
plt.show()
```
